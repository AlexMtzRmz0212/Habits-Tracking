```python
import sqlite3
import os
```

```python
def show_table_formats(db_path):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()
    cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
    tables = cursor.fetchall()
    for table_name, in tables:
        print(f"Table: {table_name}")
        cursor.execute(f"PRAGMA table_info('{table_name}')")
        columns = cursor.fetchall()
        print("Columns:")
        for col in columns:
            print(f"  {col[1]} ({col[2]})")
        print()
    conn.close()
show_table_formats('Loop Habits Backup 2025-08-28 204550.db')
```

```python
def show_first_row_each_table(db_path):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()
    cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
    tables = cursor.fetchall()
    for table_name, in tables:
        print(f"First row of table: {table_name}")
        try:
            cursor.execute(f"PRAGMA table_info('{table_name}')")
            columns = [col[1] for col in cursor.fetchall()]
            cursor.execute(f"SELECT * FROM '{table_name}' LIMIT 1")
            row = cursor.fetchone()
            if row:
                for col_name, value in zip(columns, row):
                    print(f"  {col_name}: {value}")
            else:
                print("  (No rows)")
        except Exception as e:
            print(f"  Error: {e}")
        print()
    conn.close()

show_first_row_each_table('Loop Habits Backup 2025-08-28 204550.db')
```

```python
def show_all_tables_matrix(db_path):
    conn = sqlite3.connect(db_path)
    cursor = conn.cursor()
    cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
    tables = [t[0] for t in cursor.fetchall()]
    for table_name in tables:
        print(f"Table: {table_name}")
        cursor.execute(f"PRAGMA table_info('{table_name}')")
        columns = [col[1] for col in cursor.fetchall()]
        print(" | ".join(columns))
        print("-" * (len(" | ".join(columns))))
        cursor.execute(f"SELECT * FROM '{table_name}'")
        rows = cursor.fetchall()
        for row in rows:
            print(" | ".join(str(v) for v in row))
        print()
    conn.close()

show_all_tables_matrix('Loop Habits Backup 2025-08-28 204550.db')
```
